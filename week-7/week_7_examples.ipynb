## Week 7 Example - Principal Component Analysis

```python
import pandas as pd
myData = pd.read_excel('jaggia_ba_2025_ch11_data(1).xlsx', sheet_name = 'Baseball_Players')
myData.head()
```

```python
myDataNum = myData.drop('Player', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
myDataSt = scaler.fit_transform(myDataNum)
```

```python
myDataSt = pd.DataFrame(myDataSt, columns=myDataNum.columns, index=myDataNum.index)
myDataSt.head()
```

```python
myDataSt.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pcaResults = pca.fit(myDataSt)
```

```python
expVariance = pcaResults.explained_variance_ratio_
expVariance
```

```python
import numpy as np
pcaSummary = pd.DataFrame({'Explained variance': expVariance,
                          'Cumulative variance': np.cumsum(expVariance)})
np.round(pcaSummary, decimals = 4)
```

```python
pcaWeights = pd.DataFrame(
    pcaResults.components_, 
    columns = myDataNum.columns)
np.round(pcaWeights, decimals = 4)
```

```python
pcaScores = pd.DataFrame(pcaResults.fit_transform(myDataSt))
```

```python
pcaScores.insert(0, 'Player', myData['Player'])
```

```python
np.round(pcaScores, decimals = 4)
```
